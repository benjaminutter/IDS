Here I am starting the Python part:


```python
import pandas as pd;
```

This is where I upload the data and print it

```python

link = f'https://raw.githubusercontent.com/bforoura/IDS/main/HW3/Life_Expectancy_Mac.csv';
df = pd.read_csv(link);
df.rename(columns={'Life_Expectancy ': 'Life_Expec'}, inplace=True)
print(df);
```

Data cleansing and reprint:

(population, alcohol, bmi and gdp)

```python
df['Population'] = df['Population'].fillna(df.mean(numeric_only=True));
df['Alcohol'] = df['Alcohol'].fillna(df.mean(numeric_only=True));
df['GDP'] = df['GDP'].fillna(df.mean(numeric_only=True));
df['Total_Expenditure'] = df['Total_Expenditure'].fillna(df.mean(numeric_only=True));
df['Adult_Mortality'] = df['Adult_Mortality'].fillna(df.mean(numeric_only=True));
df['Percentage_Expenditure'] = df['Percentage_Expenditure'].fillna(df.mean(numeric_only=True));


print(df);
```

How does Adult mortality rates affect life expectancy?

---



```python
import numpy as np;
x = df.Adult_Mortality;
y = df.Life_Expec;
correlation = np.corrcoef(x, y)[0, 1];

print("Correlation coefficient:", correlation);
```
