Here I am starting the Python part:


```python
import pandas as pd;
```

This is where I upload the data and print it

```python

link = f'https://raw.githubusercontent.com/bforoura/IDS/main/HW3/Life_Expectancy_Mac.csv';
df = pd.read_csv(link);
df.rename(columns={'Life_Expectancy ': 'Life_Expec'}, inplace=True)
print(df);
```

Data cleansing and reprint:

(population, alcohol, bmi and gdp)

```python
df['Total_Expenditure'] = pd.to_numeric(df['Total_Expenditure'], errors='coerce')
df['Total_Expenditure'].fillna(df['Total_Expenditure'].mean(), inplace=True)

df['Population'] = pd.to_numeric(df['Population'], errors='coerce')
df['Population'].fillna(df['Population'].mean(), inplace=True)

df['Alcohol'] = pd.to_numeric(df['Alcohol'], errors='coerce')
df['Alcohol'].fillna(df['Alcohol'].mean(), inplace=True)

df['GDP'] = pd.to_numeric(df['GDP'], errors='coerce')
df['GDP'].fillna(df['GDP'].mean(), inplace=True)

df['Adult_Mortality'] = pd.to_numeric(df['Adult_Mortality'], errors='coerce')
df['Adult_Mortality'].fillna(df['Adult_Mortality'].mean(), inplace=True)

df['Percentage_Expenditure'] = pd.to_numeric(df['Percentage_Expenditure'], errors='coerce')
df['Percentage_Expenditure'].fillna(df['Percentage_Expenditure'].mean(), inplace=True)

df['Schooling'] = pd.to_numeric(df['Schooling'], errors='coerce')
df['Schooling'].fillna(df['Schooling'].mean(), inplace=True)

df['Life_Expec'] = pd.to_numeric(df['Life_Expec'], errors='coerce')
df['Life_Expec'].fillna(df['Life_Expec'].mean(), inplace=True)

print(df)
```

How does Adult mortality rates affect life expectancy?

---



```python
import numpy as np;
x = df.Adult_Mortality;
y = df.Life_Expec;
correlation = np.corrcoef(x, y)[0, 1];

print("Correlation coefficient:", correlation);
```

Linear Regression Models:

```python
#SCHOOLING
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

x = df[['Schooling']]
y = df['Life_Expec']

#lrm
model = LinearRegression()
model.fit(x, y)

#coefficients and intercept
coef = model.coef_[0]
intercept = model.intercept_

print(f'Equation of the model: Dependent_Variable = {coef:.4f} * Independent_Variable + {intercept:.4f}')
predictions = model.predict(x)

#r2 and mse
r_squared = r2_score(y, predictions)
mse = mean_squared_error(y, predictions)

print(f'R-squared (R): {r_squared:.4f}')
print(f'Mean Squared Error (MSE): {mse:.4f}')

#graph
import matplotlib.pyplot as plt

plt.scatter(x, y, label='Data Points')
plt.plot(x, predictions, color='red', label='Linear Regression Line')
plt.xlabel('Schooling')
plt.ylabel('Life_Expec')
plt.title('Schooling on Life Expectancy')
plt.legend()
plt.show()
```
